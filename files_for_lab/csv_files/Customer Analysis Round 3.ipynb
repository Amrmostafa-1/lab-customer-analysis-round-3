```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
data
```

```python
sns.countplot(x='Response', data=data)
plt.show()
```

```python
SC_unique = data['Sales Channel'].unique()
SC_unique
```

```python
filtered_agent = data[data['Sales Channel'] == 'Agent']
filtered_cc = data[data['Sales Channel'] == 'Call Center']
filtered_web = data[data['Sales Channel'] == 'Web']
filtered_branch = data[data['Sales Channel'] == 'Branch']
filtered_branch
```

```python
filtered_agent.describe(include = [object]).T
```

```python
#we tried to go a really long way around but used chatgpt to optimze the code 
response_counts_agent = filtered_agent['Response'].value_counts(normalize=True)
yes_percentage_agent = response_counts_agent['Yes'] * 100
response_counts_cc = filtered_cc['Response'].value_counts(normalize=True)
yes_percentage_cc = response_counts_cc['Yes'] * 100
response_counts_web = filtered_web['Response'].value_counts(normalize=True)
yes_percentage_web = response_counts_web['Yes'] * 100
response_counts_branch = filtered_branch['Response'].value_counts(normalize=True)
yes_percentage_branch = response_counts_branch['Yes'] * 100
display(round(yes_percentage_agent, 2))
display(round(yes_percentage_cc, 2))
display(round(yes_percentage_web, 2))
display(round(yes_percentage_branch, 2))

yes_per = [yes_percentage_agent, yes_percentage_cc, yes_percentage_web, yes_percentage_branch]
```

```python

response_rates_Sales = data[data['Response'] == 'Yes']['Sales Channel'].value_counts() / data['Sales Channel'].value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=response_rates_Sales.index, y=response_rates_Sales.values)
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate')
plt.show()
```

```python

response_rates_TCA = data[data['Response'] == 'Yes']['Total Claim Amount'].value_counts() / data['Total Claim Amount'].value_counts()


plt.figure(figsize=(10, 6))
sns.histplot(response_rates_TCA, kde=True, label='Yes')
plt.title('Distribution of Total Claim Amount by Response')
plt.xlabel('Total Claim Amount')
plt.ylabel('Density')
plt.legend()

plt.show()
```

```python

response_rates_Income = data[data['Response'] == 'Yes']['Income'].value_counts() / data['Income'].value_counts()


plt.figure(figsize=(10, 6))
sns.histplot(response_rates_Income, kde=True, label='Yes')
plt.title('Distribution of Income by Response')
plt.xlabel('Income')
plt.ylabel('Density')
plt.legend()

plt.show()
```
